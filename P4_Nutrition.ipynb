# Utilisation des céréales 


```python
df_dispo_cereales = df_dispo_alim[df_dispo_alim["Produit"].isin(["Blé","Seigle","Orge","Avoine","Maïs","Riz (paddy)","Mélanges de céréales","Sarrasin","Sorgho","Mil","Quinoa","Autres céréales"])]
df_dispo_cereales
```

```python
df_dispo_zone_cereale = df_dispo_cereales.groupby("Zone")[["Aliments pour animaux","Nourriture","Disponibilité intérieure"]].sum()
df_dispo_zone_cereale
```

```python
df_dispo_zone_cereale["Ratio utilisation cereales Animaux"] = df_dispo_zone_cereale["Aliments pour animaux"] / df_dispo_zone_cereale["Disponibilité intérieure"]
df_dispo_zone_cereale["Ratio utilisation cereales Humains"] = df_dispo_zone_cereale["Nourriture"] / df_dispo_zone_cereale["Disponibilité intérieure"]
df_dispo_zone_cereale
```

```python
df_global_repartition_cereales = df_dispo_zone_cereale[["Ratio utilisation cereales Animaux","Ratio utilisation cereales Humains"]].mean()
df_global_repartition_cereales
```

```python
df_global_repartition_cereales["Autres"]=1-(df_global_repartition_cereales["Ratio utilisation cereales Animaux"]+df_global_repartition_cereales["Ratio utilisation cereales Humains"])
df_global_repartition_cereales
```

```python
type(df_global_repartition_cereales)
```

```python
plot = df_global_repartition_cereales.plot.pie( figsize=(10, 30), title="Répartition des Céréales", legend=True, \
                   autopct='%1.1f%%', \
                   shadow=False, startangle=0)
```
